### Create a tensor

```
hello = tf.constant('Hello World')
```

```
print(hello)
```

### Tensor name and shape

```
hello_2 =  tf.constant('Hello World', shape=[6,2], name='my_tensor')
print(hello_2)
```

### Adding two tensors

```
a = tf.constant([1.2])
b = tf.constant([3.4])
c = tf.add(a,b)
print(c)
```

### How do we get tensor Output

```
#Start a session
sess = tf.Session()
```

```
#Execute hello
print(sess.run(hello))
```

```
#Execute hello_2
print(sess.run(hello_2))
```

```
#Adding a and b
sess.run(c)
```

```
sess.run([c,hello, hello_2])
```

Close the session

```
sess.close()
```

```
with tf.Session() as sess:
    print(sess.run(hello))
```
