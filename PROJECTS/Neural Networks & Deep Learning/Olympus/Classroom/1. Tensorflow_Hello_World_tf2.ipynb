### Create a tensor

```
hello = tf.constant('Hello World')
```

```
print(hello)
```

```
print(hello.numpy())
```

### Tensor name and shape

```
hello_2 =  tf.constant('Hello World', shape=[6,2])
print(hello_2.numpy())
```

### Adding two tensors

```
a = tf.constant([1.2])
b = tf.constant([3.4])
c = tf.add(a,b)
print(c.numpy())
```
